# Migraine Tracker & Predictor
### Q-Phase (QUA³CK-Prozessmodell)


## 1. Problemstellung

Migräne ist eine neurologische Erkrankung, deren Auftreten stark von individuellen Faktoren wie Schlaf, Stress, Ernährung und Umweltbedingungen abhängt.

Für Betroffene ist es schwierig, persönliche Trigger zu erkennen und Migräneattacken frühzeitig vorherzusagen.

Dieses Projekt zielt darauf ab, ein Machine-Learning-Modell zu entwickeln, das basierend auf verschiedenen Einflussfaktoren die Wahrscheinlichkeit einer Migräneattacke vorhersagt.


## 2. Zielgruppe

- Personen mit regelmäßiger Migräne
- Nutzer von Gesundheits-Tracking-Apps
- langfristig: medizinische Analyse / Prävention


## 3. Ziel der Anwendung

Die geplante Anwendung soll:

- persönliche Gesundheits- und Lifestyle-Daten erfassen
- externe Daten (z. B. Wetter) integrieren
- Muster zwischen Einflussfaktoren und Migräne erkennen
- eine tägliche Risikowahrscheinlichkeit berechnen

Beispiel:
"Heute besteht eine 72% Wahrscheinlichkeit für Migräne aufgrund von wenig Schlaf, hohem Stress und niedrigem Luftdruck."


## 4. ML-Fragestellung

Zentrale Frage:

Wie hoch ist die Wahrscheinlichkeit für eine Migräneattacke basierend auf individuellen und externen Faktoren?

Input (Features):
- Schlafdauer
- Stresslevel
- Bildschirmzeit
- Ernährung
- Wetterdaten (Temperatur, Luftdruck, Luftfeuchtigkeit, Niederschlag)

Output:
- Wahrscheinlichkeit für Migräne (0–100%)


## 5. Datenquellen

Das Projekt basiert auf mehreren Datenquellen (Big-Data-Ansatz):

1. Öffentliche Datensätze (Kaggle, OpenML)
   - Migraine / Headache Datasets
   - Sleep & Lifestyle Datasets

2. Wetterdaten
   - Temperatur
   - Luftdruck
   - Luftfeuchtigkeit
   - Niederschlag

3. Eigene Daten (optional)
   - persönliches Migräne-Tagebuch
   - Schlaf, Stress, Aktivität

Ziel ist die Kombination dieser Datenquellen zu einem einheitlichen Datensatz.


## 6. Geplante Features

| Kategorie | Features |
|----------|--------|
| Lifestyle | Schlafdauer, Stress, Bildschirmzeit |
| Gesundheit | Migräne (ja/nein), Intensität |
| Wetter | Temperatur, Luftdruck, Luftfeuchtigkeit |
| Kontext | Wochentag, Aktivität |


## 7. Erfolgsmetriken (KPIs)

- Accuracy / F1-Score des Modells
- Recall für Migräne-Fälle (wichtig)
- Interpretierbarkeit (Feature Importance)
- Performance der Vorhersage (< 1 Sekunde)


## 8. Deployment-Ziel

- Entwicklung einer interaktiven Streamlit-Web-App
- Nutzer können eigene Daten eingeben
- Anzeige von:
  - Risiko-Wahrscheinlichkeit
  - wichtigsten Einflussfaktoren
  - Handlungsempfehlungen


## 9. Motivation für Machine Learning

Die Zusammenhänge zwischen Einflussfaktoren und Migräne sind komplex und individuell unterschiedlich.

Machine Learning ermöglicht:
- Erkennung nicht-linearer Zusammenhänge
- Personalisierung
- kontinuierliche Verbesserung durch neue Daten

Damit geht das Projekt über eine einfache Tracking-App hinaus.
