# Migraine Tracker & Predictor
### Q-Phase (QUA³CK-Prozessmodell)


## Problemstellung

Migräne wird in der Praxis häufig in Form von Tagebüchern dokumentiert. Nutzer erfassen rückblickend, wann eine Attacke aufgetreten ist und welche Symptome vorlagen. Diese Daten sind zwar wertvoll, ermöglichen jedoch nur begrenzt eine aktive Prävention. Für Betroffene ist es schwierig, persönliche Trigger zu erkennen und Migräneattacken frühzeitig vorherzusagen.

Gleichzeitig ist bekannt, dass Migräne nicht durch einzelne Faktoren entsteht, sondern durch komplexe Wechselwirkungen aus Lifestyle-, Umwelt- und physiologischen Einflüssen. Faktoren wie Schlafmangel, Stress, Hydration oder Wetterveränderungen können individuell unterschiedliche Auswirkungen haben.

Genau hier setzt der Migraine Tracker & Predictor (Arbeitsname "**MigraCast**") an. Das Projekt verfolgt das Ziel, diese verschiedenen Einflussfaktoren mithilfe von Machine Learning zu kombinieren und daraus eine **individuelle Migräne-Wahrscheinlichkeit** abzuleiten. Anstatt Migräne nur rückblickend zu dokumentieren, soll die Anwendung eine **proaktive, datenbasierte Einschätzung zukünftiger Risiken** ermöglichen.


---

## 🔬 Zentrale Forschungsfrage

**Wie lassen sich persönliche Lifestyle-Daten, Gesundheitsdaten und externe Umweltfaktoren wie Wetter so modellieren und kombinieren, dass daraus eine individuelle Wahrscheinlichkeit für das Auftreten von Migräne berechnet und für Nutzer verständlich dargestellt werden kann?**

---

## 🧭 Thematische Forschungsfragen

### 1. Einflussfaktoren identifizieren

Welche Faktoren wie Schlaf, Stress, Hydration, Bildschirmzeit und Wetterbedingungen stehen in Zusammenhang mit dem Auftreten von Migräne?

---

### 2. Modellierung der Migräne-Wahrscheinlichkeit

Wie kann ein Machine-Learning-Modell aus diesen Faktoren eine tägliche Wahrscheinlichkeit für das Auftreten einer Migräneattacke berechnen?

---

### 3. Verständliche Visualisierung und Nutzbarkeit

Wie lassen sich die Modelloutputs so darstellen, dass Nutzer ihre individuellen Trigger erkennen und daraus konkrete Maßnahmen ableiten können?

---


## 🎯 Zielgruppe und Anwendungsnutzen

Die Anwendung richtet sich primär an:

- Personen mit regelmäßiger Migräne  
- Nutzer von Gesundheits- und Tracking-Apps  
- langfristig auch medizinisch Interessierte  

Der Mehrwert liegt darin, Migräne nicht nur zu dokumentieren, sondern:

- persönliche Gesundheits- und Lifestyle-Daten erfassen
- individuelle Trigger sichtbar zu machen  
- externe Daten (z. B. Wetter) integrieren
- Zusammenhänge zwischen Verhalten und Migräne zu erkennen  
- eine tägliche Risikoabschätzung zu erhalten  
- präventive Maßnahmen zu ermöglichen  

**Beispiel:**

„Heute erhöhtes Risiko (72 %) aufgrund von wenig Schlaf, hohem Stress und Luftdruckveränderungen.“

## 🗂️ Datengrundlage

### 1. Hauptdatensatz

**Migraine Dataset from Wearable Devices**

Dieser Datensatz enthält:

- sleep_hours  
- mood_level  
- stress_level  
- hydration_level  
- screen_time  
- migraine_occurrence  
- migraine_severity  

Die Zielvariable ist:

```migraine_occurrence```


---

### 2. Lifestyle-Datensatz

**Sleep Health and Lifestyle Dataset**

Zusätzliche Features:

- Sleep Duration  
- Quality of Sleep  
- Physical Activity Level  
- Stress Level  
- Heart Rate  
- Daily Steps  

Dieser Datensatz erweitert die Featurebasis.

---

### 3. Wetterdaten

**Open-Meteo API**

Verwendete Variablen:

- Temperatur  
- Luftdruck  
- Luftfeuchtigkeit  
- Niederschlag  

Diese Daten ermöglichen die Integration externer Einflussfaktoren.

---

### Datenstrategie

Der Hauptdatensatz liefert die Zielvariable und erste Features.  
Lifestyle- und Wetterdaten erweitern das Modell zu einem **multimodalen Ansatz**, der interne und externe Einflussfaktoren kombiniert.

---


## 🤖 Machine-Learning-Ansatz

Zentrale Frage:

Wie hoch ist die Wahrscheinlichkeit für eine Migräneattacke basierend auf individuellen und externen Faktoren?

Input (Features):
- Schlafdauer
- Stresslevel
- Bildschirmzeit
- Ernährung
- Wetterdaten (Temperatur, Luftdruck, Luftfeuchtigkeit, Niederschlag)


Mögliche Modelle:
- Logistic Regression  
- Random Forest   

Output:
- Wahrscheinlichkeit für Migräne (0–100%)

---

## ⚖️ Grenzen und Schwächen

- Die App stellt **keine medizinische Diagnose** dar.  
- Vorhersagen basieren auf Wahrscheinlichkeiten.  
- Datensätze sind teilweise generisch oder synthetisch.  
- Individuelle Unterschiede können nur begrenzt abgebildet werden.  
- Personalisierung erfordert langfristig eigene Nutzerdaten.  
- Machine Learning basiert auf historischen Daten und kann neue/extreme Ereignisse nur eingeschränkt erfassen.  

---


## ⚖️ Abgrenzung

Viele bestehende Anwendungen im Bereich Migräne sind reine Tracking-Apps.

**MigraCast unterscheidet sich dadurch, dass:**

- nicht nur Daten gesammelt werden  
- sondern eine Vorhersage getroffen wird  
- individuelle Einflussfaktoren sichtbar gemacht werden  

---

## 📊 Erweiterte Datenstrategie: Eigene Nutzerdaten

Neben öffentlichen Datensätzen spielt die Integration eigener Nutzerdaten eine zentrale Rolle für das Projekt.

Die App soll es ermöglichen, dass Nutzer täglich persönliche Daten erfassen, z. B.:

- Schlafdauer
- Stresslevel
- Hydration
- Bildschirmzeit
- Stimmung
- Auftreten von Migräne (ja/nein)

Diese Daten werden im Verlauf gespeichert und bilden eine individuelle Zeitreihe.

### Ziel der eigenen Daten

Während die öffentlichen Datensätze als Trainingsbasis dienen, ermöglichen eigene Daten:

- Personalisierung des Modells  
- Anpassung an individuelle Trigger  
- kontinuierliches Lernen  

Langfristig kann das Modell dadurch erkennen:

- welche Faktoren für einen spezifischen Nutzer besonders relevant sind  
- welche Kombinationen von Faktoren Migräne begünstigen  
- welche Muster sich über Zeit entwickeln  

Beispiel:

Ein Nutzer reagiert besonders stark auf:
- wenig Schlaf + hoher Stress  

Ein anderer Nutzer reagiert eher auf:
- Luftdruckveränderungen  

Die App kann diese Unterschiede erkennen und entsprechend gewichten.

---

## 🧠 Nutzung der Daten in der App

Die Anwendung kombiniert:

- öffentliche Trainingsdaten (generelles Modell)
- individuelle Nutzerdaten (Personalisierung)

Daraus ergeben sich zwei Ebenen:

### 1. Globales Modell

- basiert auf allgemeinen Datensätzen  
- erkennt generelle Zusammenhänge  
- liefert initiale Vorhersagen  

### 2. Individuelle Anpassung (zukünftig)

- nutzt persönliche Verlaufsdaten  
- passt Gewichtung der Features an  
- verbessert Vorhersagegenauigkeit für den einzelnen Nutzer  

Ziel ist ein System, das mit der Zeit **individuell lernt**.
